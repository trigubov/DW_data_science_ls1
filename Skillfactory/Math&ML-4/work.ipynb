# ОПЕРАЦИИ НАД МНОЖЕСТВАМИ В PYTHON

```python
A = {'ID453', 'ID312', 'ID41', 'ID3', 'ID500', 'ID920', 'ID36', 'ID27'}
B = {'ID41', 'ID36', 'ID27', 'ID124', 'ID7', 'ID501', 'ID91' }
```

```python
union_AB = A.union(B)
print(union_AB)
#{'ID500', 'ID27', 'ID41', 'ID3', 'ID501', 'ID453', 'ID312', 'ID124', 'ID920', 'ID91', 'ID7', 'ID36'}
```

```python
# Метод union() можно заменить простой вертикальной чертой:

union_AB = A | B
print(union_AB)
#{'ID500', 'ID27', 'ID41', 'ID3', 'ID501', 'ID453', 'ID312', 'ID124', 'ID920', 'ID91', 'ID7', 'ID36'}
```

## ПЕРЕСЕЧЕНИЕ МНОЖЕСТВ

```python
inter_AB = A.intersection(B)
print(inter_AB)
#{'ID27', 'ID41', 'ID36'}
```

```python
# Также мы можем заменить метод intersection() на амперсанд (&) — результат будет идентичным:

inter_AB = A & B
print(inter_AB)
#{'ID27', 'ID41', 'ID36'}
```

## РАЗНОСТЬ МНОЖЕСТВ

```python
diff_AB = A.difference(B)
print(diff_AB)
#{'ID500', 'ID3', 'ID453', 'ID312', 'ID920'}
```

```python
diff_AB = A - B
print(diff_AB)
#{'ID500', 'ID3', 'ID453', 'ID312', 'ID920'}
```

```python
diff_BA = B - A
print(diff_BA)
#{'ID124', 'ID7', 'ID501', 'ID91'}
```

## СИММЕТРИЧЕСКАЯ РАЗНОСТЬ

```python
symmAB = A.symmetric_difference(B)
print(symmAB)
#{ID124', 'ID91', 'ID7', 'ID312', 'ID500', 'ID453', 'ID3', 'ID501', 'ID920'}


symmAB = A ^ B
print(symmAB)
#{ID124', 'ID91', 'ID7', 'ID312', 'ID500', 'ID453', 'ID3', 'ID501', 'ID920'}
```

## Найти экстремум функции

```python
from sympy import symbols, diff, solve

# Определение переменных
x, y = symbols('x y')

# Определение функции
z = 4*x**2 - 6*x*y - 34*x + 5*y**2 + 42*y + 7

# Находим частные производные по x и y
dz_dx = diff(z, x)
dz_dy = diff(z, y)

# Находим точки, где производные равны нулю
critical_points = solve((dz_dx, dz_dy), (x, y))
print("Точки, где производные равны нулю:", critical_points)
```

```python
import numpy as np
 
def func_rosen(x):
    r = np.sum(100*(x[1:]-x[:-1]**2)**2 + (1-x[:-1])**2, axis=0)
    return r
```

```python
import matplotlib.pyplot as plt
graph = plt.figure(figsize=[25, 15])
ax = graph.add_subplot(projection='3d')
ax.view_init(45, 30)
x1 = np.arange(-3, 3, 0.2)
x2 = np.arange(-2, 4, 0.2)
x1, x2 = np.meshgrid(x1, x2)
y = func_rosen(np.array([x1,x2]))
surf = ax.plot_surface(x1, x2, y)
plt.show()
```

```python
from scipy.optimize import minimize
import numpy as np
x1 = np.array([2.4, 1.5, 2.1, 0.7, 1.1])
result_point = minimize(func_rosen, x1, method='powell')
print(result_point.x)
```

```python
from scipy.optimize import minimize
import numpy as np
x1 = np.array([2.4, 1.5, 2.1, 0.7, 1.1])
result_point = minimize(func_rosen, x1, method='nelder-mead')
print(result_point.x)
```

# Функция Лагранджа


```python
from sympy import *
a,b,w=symbols(' a b w' )
g = 3*a**2 + a + 4*b + 5*b**2
print('Целевая функция для аргументов a и b :\n f = ', g)
q = a + b - 200
print('Функция ограничений: ', q,'= 0')
f = 3*a**2 + a + 4*b + 5*b**2 + w*(a + b - 200)
print('Функция Лагранжа :\n ',f)
fa = f.diff(a)
print('df/da =',fa,'= 0')
fb = f.diff(b)
print('df/db =',fb,'= 0')
fw = f.diff(w)
print('df/dw =',fw,'= 0')
sols = solve([fa,fb,fw],a,b,w)
print('Стационарная точка M(x,y):\n',float(sols[a]),',',float(sols[b]))

#Целевая функция для аргументов a и b :
#f =  3*a**2 + a + 5*b**2 + 4*b
#Функция ограничений:  a + b - 200 = 0
#Функция Лагранжа :
#3*a**2 + a + 5*b**2 + 4*b + w*(a + b - 200)
#df/da = 6*a + w + 1 = 0
#df/db = 10*b + w + 4 = 0
#df/dw = a + b - 200 = 0
#Стационарная точка M(x,y):
#125.1875 , 74.8125
```

```python
# Импортируем все необходимые функции из библиотеки SymPy
from sympy import init_printing, solve, symbols

# Подключаем аккуратный вывод формул
init_printing()

# Создаём переменные
x1, x2, x1_tilde, lambda1, lambda2 = symbols(
    "x_1 x_2 xtilde_2 \lambda_1 \lambda_2", real=True
)

# Пишем систему уравнений для переменных
dLdx1 = -x2 + lambda1 - lambda2
dLdx2 = -x1 + lambda1
dLdx1_tilde = 2 * lambda2 * x1_tilde
dLdlambda1 = x1 + x2 - 10
dLdlambda2 = -x1 + 6 + x1_tilde**2

# Решаем систему уравнений с помощью функции solve
solution = solve(
    [dLdx1, dLdx2, dLdx1_tilde, dLdlambda1, dLdlambda2],
    x1,
    x2,
    x1_tilde,
    lambda1,
    lambda2,
)

print(solution)
```

# Градиентный спуск

```python
import numpy as np

def fun(x, y, a=1, b=1):
    return a * (x ** 2) + b * (y ** 2)
```

```python
def grad(x, y, a=1, b=1):
    return np.array([2 * a * x, 2 * b * y])
```

```python
def grad_descend(grad, step_size=0.2, num_steps=30):
    lst = []
    x = np.random.uniform(0, 3, size = 2)
    lst.append(x)
    for i in range(num_steps):
        x = x - step_size * grad(lst[-1][0], lst[-1][1])
        lst.append(x)
    return np.array(lst)
```

```python
def plot_grad(fun, trace):
    fig = plt.figure(figsize=(10, 8))
    x_ = np.linspace(-1, 1, 100)
    y_ = np.linspace(-1, 1, 100)
    x, y = np.meshgrid(x_, y_)
    z = fun(x, y)
    ax = fig.add_subplot(1, 1, 1, projection='3d')
    ax.plot_surface(x, y, z, alpha=0.6)
    ax.contour(x, y, z, zdir='z', offset=z.min())
    z_trace = fun(trace[:, 0], trace[:, 1])
    ax.plot(trace[:, 0], trace[:, 1], z_trace, "o-")
    ax.set_xlim(x.min(), x.max())
    ax.set_ylim(y.min(), y.max())
    ax.set_zlim(z.min(), z.max())
```

```python
import matplotlib.pyplot as plt
trace = grad_descend(grad, 0.1, 20)
plot_grad(fun, trace = trace)
```
