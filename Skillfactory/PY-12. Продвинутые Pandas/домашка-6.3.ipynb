```python
import pandas as pd
import os
```

```python
def concat_user_files(path):
    filelist=os.listdir(path)
    filelist.sort()
    lst = []
    for filename in filelist:
        fullfilename = path+'\\'+filename
        lst.append(pd.read_csv(fullfilename, sep=','))
    users = pd.concat(
        lst,
        ignore_index=True
        )
    users = users.drop_duplicates()
    return users

```
