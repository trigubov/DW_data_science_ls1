```python
import pandas as pd
df = pd.read_csv('data/test_data.csv')
df
```

```python
# Если признак имеет больше 50 % пропущенных значений, удалите его.
df = df.dropna(thresh=df.shape[0]*0.5, axis=1)
df
```

```python
# Для оставшихся данных: если в строке более двух пропусков, удалите строку.
#удаляем записи, в которых есть хотя бы 1 пропуск
#thresh=df.shape[1]-2
df = df.dropna(thresh=df.shape[1]-2, axis=0)
df
```

```python
#создаём словарь 'имя_столбца': число (признак), на который надо заменить пропуски 
values = {    
    'one': df['one'].mean(),
    'two': df['two'].mean(),
    'three': df['three'].mean(),
    'four': df['four'].mode()[0]    
}
#заполняем оставшиеся записи константами в соответствии со словарем values
df = df.fillna(values)

df
```
